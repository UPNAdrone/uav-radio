# Example: processing a log from the Digital Twin
In this example, we will extract a log recorded using the Gazebo + Ardupilot digital twin, including our own custom integration of UAVRadio to simulate radio signal Path Loss (PL) using various PL models and frequencies.

```python
import matplotlib.pyplot as plt
```

```python
# Initialize empty lists to store data from the log file
times = []
distances = []
spatial_loss_fspl_868 = []
spatial_loss_alpha_beta_868 = []
spatial_loss_two_ray_868 = []
spatial_loss_modified_two_ray_868 = []
spatial_loss_dual_slope_868 = []
spatial_loss_fspl_2_4 = []
spatial_loss_alpha_beta_2_4 = []
spatial_loss_two_ray_2_4 = []
spatial_loss_modified_two_ray_2_4 = []
spatial_loss_dual_slope_2_4 = []
spatial_loss_fspl_5_8 = []
spatial_loss_alpha_beta_5_8 = []
spatial_loss_two_ray_5_8 = []
spatial_loss_modified_two_ray_5_8 = []
spatial_loss_dual_slope_5_8 = []

# Read data from the log file
with open("rf_data_full.log", "r") as log_file:
    for line in log_file:
        parts = line.strip().split(',')
        times.append(float(parts[0]))
        distances.append(float(parts[1]))
        spatial_loss_fspl_868.append(float(parts[2]))
        spatial_loss_alpha_beta_868.append(float(parts[3]))
        spatial_loss_two_ray_868.append(float(parts[4]))
        spatial_loss_modified_two_ray_868.append(float(parts[5]))
        spatial_loss_dual_slope_868.append(float(parts[6]))
        spatial_loss_fspl_2_4.append(float(parts[7]))
        spatial_loss_alpha_beta_2_4.append(float(parts[8]))
        spatial_loss_two_ray_2_4.append(float(parts[9]))
        spatial_loss_modified_two_ray_2_4.append(float(parts[10]))
        spatial_loss_dual_slope_2_4.append(float(parts[11]))
        spatial_loss_fspl_5_8.append(float(parts[12]))
        spatial_loss_alpha_beta_5_8.append(float(parts[13]))
        spatial_loss_two_ray_5_8.append(float(parts[14]))
        spatial_loss_modified_two_ray_5_8.append(float(parts[15]))
        spatial_loss_dual_slope_5_8.append(float(parts[16]))
```

```python
# Set the font size for the entire plot
plt.rcParams.update({'font.size': 13})

# Create subplots in a 2x3 grid
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 10), sharex=False, sharey=True)

# Plot for 868 MHz frequencies
axes[0, 0].plot(times, spatial_loss_fspl_868, label='FSPL')
axes[0, 0].plot(times, spatial_loss_alpha_beta_868, '--', label='Alpha-Beta')
axes[0, 0].plot(times, spatial_loss_two_ray_868, ':', label='Two Ray')
axes[0, 0].plot(times, spatial_loss_modified_two_ray_868, '-.', label='Modified Two Ray')
axes[0, 0].plot(times, spatial_loss_dual_slope_868, 'gray', label='Dual Slope')
axes[0, 0].set_title('868 MHz')
axes[0, 0].set_ylabel('Path Loss (dB)', fontsize=14)  # Increased font size
axes[0, 0].set_xlabel('\n(a)', fontsize=14)  # Increased font size
axes[0, 0].legend().set_visible(False)
axes[0, 0].set_ylim(20, 150)  # Set a common y-axis limit
axes[0, 0].grid(True, linestyle='--', alpha=0.7)  # Add grid lines

# Plot for 2.4 GHz frequencies
axes[0, 1].plot(times, spatial_loss_fspl_2_4, label='FSPL')
axes[0, 1].plot(times, spatial_loss_alpha_beta_2_4, '--', label='Alpha-Beta')
axes[0, 1].plot(times, spatial_loss_two_ray_2_4, ':', label='Two Ray')
axes[0, 1].plot(times, spatial_loss_modified_two_ray_2_4, '-.', label='Modified Two Ray')
axes[0, 1].plot(times, spatial_loss_dual_slope_2_4, 'gray', label='Dual Slope')
axes[0, 1].set_title('2.4 GHz')
axes[0, 1].set_xlabel('Time (s)\n(b)', fontsize=14)  # Increased font size
axes[0, 1].legend().set_visible(False)
axes[0, 1].set_ylim(20, 150)  # Set a common y-axis limit
axes[0, 1].grid(True, linestyle='--', alpha=0.7)  # Add grid lines

# Plot for 5.8 GHz frequencies
axes[0, 2].plot(times, spatial_loss_fspl_5_8, label='FSPL')
axes[0, 2].plot(times, spatial_loss_alpha_beta_5_8, '--', label='Alpha-Beta')
axes[0, 2].plot(times, spatial_loss_two_ray_5_8, ':', label='Two Ray')
axes[0, 2].plot(times, spatial_loss_modified_two_ray_5_8, '-.', label='Modified Two Ray')
axes[0, 2].plot(times, spatial_loss_dual_slope_5_8, 'gray', label='Dual Slope')
axes[0, 2].set_title('5.8 GHz')
axes[0, 2].set_xlabel('\n(c)', fontsize=14)  # Increased font size
axes[0, 2].legend().set_visible(False)
axes[0, 2].set_ylim(20, 150)  # Set a common y-axis limit
axes[0, 2].grid(True, linestyle='--', alpha=0.7)  # Add grid lines

# Plot for 868 MHz frequencies
axes[1, 0].plot(distances, spatial_loss_fspl_868, label='FSPL')
axes[1, 0].plot(distances, spatial_loss_alpha_beta_868, '--', label='Alpha-Beta')
axes[1, 0].plot(distances, spatial_loss_two_ray_868, ':', label='Two Ray')
axes[1, 0].plot(distances, spatial_loss_modified_two_ray_868, '-.', label='Modified Two Ray')
axes[1, 0].plot(distances, spatial_loss_dual_slope_868, 'gray', label='Dual Slope')
axes[1, 0].set_ylabel('Path Loss (dB)', fontsize=14)  # Increased font size
axes[1, 0].set_xlabel('\n(d)', fontsize=14)  # Increased font size
axes[1, 0].legend().set_visible(False)
axes[1, 0].set_ylim(20, 150)  # Set a common y-axis limit
axes[1, 0].grid(True, linestyle='--', alpha=0.7)  # Add grid lines

# Plot for 2.4 GHz frequencies
axes[1, 1].plot(distances, spatial_loss_fspl_2_4, label='FSPL')
axes[1, 1].plot(distances, spatial_loss_alpha_beta_2_4, '--', label='Alpha-Beta')
axes[1, 1].plot(distances, spatial_loss_two_ray_2_4, ':', label='Two Ray')
axes[1, 1].plot(distances, spatial_loss_modified_two_ray_2_4, '-.', label='Modified Two Ray')
axes[1, 1].plot(distances, spatial_loss_dual_slope_2_4, 'gray', label='Dual Slope')
axes[1, 1].set_xlabel('Distance (m)\n(e)', fontsize=14)  # Increased font size
axes[1, 1].legend().set_visible(False)
axes[1, 1].set_ylim(20, 150)  # Set a common y-axis limit
axes[1, 1].grid(True, linestyle='--', alpha=0.7)  # Add grid lines

# Plot for 5.8 GHz frequencies
axes[1, 2].plot(distances, spatial_loss_fspl_5_8, label='Free Space PL')
axes[1, 2].plot(distances, spatial_loss_alpha_beta_5_8, '--', label='Alpha-Beta PL')
axes[1, 2].plot(distances, spatial_loss_two_ray_5_8, ':', label='Two Ray PL')
axes[1, 2].plot(distances, spatial_loss_modified_two_ray_5_8, '-.', label='Modified Two Ray PL')
axes[1, 2].plot(distances, spatial_loss_dual_slope_5_8, 'gray', label='Dual Slope PL')
axes[1, 2].set_xlabel('\n(f)', fontsize=14)  # Increased font size
axes[1, 2].legend()
axes[1, 2].set_ylim(20, 150)  # Set a common y-axis limit
axes[1, 2].grid(True, linestyle='--', alpha=0.7)  # Add grid lines

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.savefig('path_loss_dt_comparison.svg', format='svg', bbox_inches='tight')
# Show the plots
plt.show()
```
